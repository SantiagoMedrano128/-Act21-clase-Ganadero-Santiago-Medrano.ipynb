```python
# Definición de la clase
class Ganadero:
    def __init__(self, nombre, telefono, edad):
        # Atributos de la clase
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    # Función 1: Arrear
    def arrear(self):
        print(f"{self.nombre} está arreando al ganado hacia el corral.")

    # Función 2: Alimentar
    def alimentar(self):
        print(f"{self.nombre} está repartiendo el forraje para los animales.")

    # Función 3: Cuidar ganado
    def cuidar_ganado(self):
        print(f"{self.nombre} está revisando que el ganado esté sano y seguro.")

# --- Creación de Objetos (Instancias) ---

# Objeto 1: Ganadero1
ganadero1 = Ganadero("Juan Pérez", "555-0123", 45)

# Objeto 2: Ganadero2
ganadero2 = Ganadero("María López", "555-0987", 38)

# --- Uso de los objetos ---

print(f"--- Ficha del Ganadero 1 ---")
print(f"Nombre: {ganadero1.nombre}, Tel: {ganadero1.telefono}, Edad: {ganadero1.edad}")
ganadero1.arrear()
ganadero1.alimentar()

print("\n--- Ficha del Ganadero 2 ---")
print(f"Nombre: {ganadero2.nombre}, Tel: {ganadero2.telefono}, Edad: {ganadero2.edad}")
ganadero2.cuidar_ganado()
ganadero2.alimentar()

print("\n" + "-"*30 + "\n")
print("Programa terminado")
print("Creador de software por: Santiago Medrano NL = 37")
```
